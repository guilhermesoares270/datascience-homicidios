```python
import pandas as pd
pd.options.display.float_format = '${:,.2f}'.format
global df
df = pd.read_csv('dataset.csv')
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

ax = plt.gca()
localidade = 'Capital'

localidades = df.loc[df['Localidade'] == localidade]
localidades.plot.bar(
    x='Ano', y='Quantidade de OCORRENCIAS por Homicidio Doloso', rot=45, ax=ax)

ocorrencias = localidades['Quantidade de OCORRENCIAS por Homicidio Doloso'].to_numpy()
media = ocorrencias.mean()
variancia = ocorrencias.var()
mediana = np.median(ocorrencias)
moda = stats.mode(ocorrencias)

desvio_padrao = statistics.stdev(ocorrencias)
print(f'média: {round(media, 2)}')
print(f'variancia: {round(variancia, 2)}')
print(f'desvio padrão: {round(desvio_padrao, 2)}')
print(f'mediana: {round(mediana, 2)}')
#print(f'moda: {moda}')
```
